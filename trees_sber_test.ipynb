```python
import pptree

class Node(object):

  """
  Класс узлов
  """

  def __init__(self, name='root',
               children=None, parent=None):

    self.name = name
    self.parent = parent
    self.children = []
    #self.level = self.get_level()

    if children is not None:
      for child in children:
        self.add_child(child)

  def add_child(self, node):
    """
    Добавление потомков как детей
    """
    node.parent = self
    assert isinstance(node, Node)

    self.children.append(node)

  def is_root(self):
    if self.parent is None:
      return True
    else:
      return False

  def is_leaf(self):
    if len(self.children) == 0:
      return True
    else:
      return False

  def depth(self):
    """
    Рекурсивная функция, ищущая глубину, на котором сидит узел
    """
    if self.is_root():
      return 0
    else:
      return 1 + self.parent.depth()

  def disp(self):
    pptree.print_tree(self, 'children', 'name')


class Tree():
  """
  Класс как сборник узлов (что и является деревом)
  """
  def __init__(self):

    self.root = None
    self.nodes = []

  def insert(self, node, parent=None):
    """
    Назначение связи между двумя уже существующими узлами
    """

    if parent is not None:
      parent.add_child(node)

    elif self.root is None:
      self.root = node

    self.nodes.append(node)

  def new_insert(self, data, parent):
    """
    Назначение связи между двумя узлами при добавлении нового
    """
    node = Node(data)
    if parent is not None:
      parent.add_child(node)

    elif self.root is None:
      self.root = node

    self.nodes.append(node)

  def get_index(self, data):
    """
    Получает имя узла и возвращает его индекс в self.nodes
    """
    index = -1
    for N in self.nodes:
      index += 1
      if N.name == data:
        break
    if index == len(self.nodes) - 1:
      return -1
    else:
      return index

  def get_root(self):
    return self.root

  def get_node(self, id):
    return self.nodes[id]

  def get_depth(self,p: Node):
        return p.depth()

  def get_height(self, p : Node):
    """
    Глубина поддерева с корнем в p
    """
    if p.is_leaf():
      return 0
    else:
      names = []
      for child in p.children:
        names.append(child.name)
      return 1 + max(self.get_height(c) for c in p.children)

  def node_is(self, object_node, parent : Node):

    if type(object_node) is list:
      self.node_is_list(object_node, parent)

    elif type(object_node) is int:
      self.node_is_int(object_node, parent)


  def node_is_list(self, object_node, parent):

    for node in object_node:

      mid_node = Node(node)
      self.node_is(node, mid_node)



  def node_is_int(self, object_node, parent : Node):

    sub_node = Node(object_node)
    self.insert(sub_node, parent)



  def unpack_frame_list(self, frame, parent : Node):

    print(f"parent.name in call func -- {parent.name}")
    if type(frame) is list:

      for data_key in frame:
        """
        для листа отдельная функция
        """

        print(f"frame is{frame}\n")
        print(f"\n иду вглубь по {data_key}")
        if type(data_key) is int:
          data_key_node = Node(data_key)
          self.insert(data_key_node, parent)
          print(f"{data_key} is int and son of {parent.name} \n")
          #self.root.disp()

        elif type(data_key) is list:
          sub_node = Node(data_key)
          self.insert(sub_node, parent)

          for sub_sub in data_key:

            self.unpack_frame_list(sub_sub, sub_node) #как-то рекурсию завести

            print(f"data_key is {type(data_key)}, sub_node is {sub_node.name}, sub_sub is {sub_sub}")
            print(f"in list check sub sub in data_key is {sub_sub}")
            print(f"in list check сын {sub_sub}, отец {sub_node.name}")

            #self.root.disp()

    elif type(frame) is int and frame != parent.name:
        int_node = Node(frame)
        print(f"ребеночек инт -- {int_node.name}")
        print(f"батя инта -- {parent.name} \n")
        self.insert(int_node, parent)

```
